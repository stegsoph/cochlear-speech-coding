```julia
using PyPlot
using WAV
using DSP
using FFTW
```

```julia
path = "./data/"
file = "PR-open_front_unrounded_vowel.wav"
#file = "add_the_sum.wav"
wav_file = wavread(string(path,file));
audio = wav_file[1][:,1]; # just one channel
Fs = wav_file[2]; # sampling rate
time = (0:length(audio)-1)/Fs;
```

```julia
figure(figsize=(12,3))

plot(time, audio)
xlim([first(time),last(time)])
xlabel("Time in s") 
ylabel("Amplitude") 
title(file)
```

```julia
function plot_spectrogram(s, fs; t_window::Float64 = 25e-3, t_overlap ::Float64=10e-3)
    S = spectrogram(s[:,1], convert(Int, round(t_window*fs)),
                    convert(Int, round(t_overlap*fs)); window=hanning)
    t = S.time
    f = S.freq
    imshow(reverse(log10.(S.power), dims = 1), extent=[first(t)/fs, last(t)/fs,
             fs*first(f), fs*last(f)], aspect="auto", cmap="jet", vmin=1, vmax=-15)
    xlabel("Time in s")
    ylabel("Frequency in Hz")
    cbar = colorbar()
    cbar.set_label("dB")
end;
```

```julia
audio_fft = fft(audio)
audio_fft = FFTW.fftshift(audio_fft)

freqs = FFTW.fftfreq(length(time), Fs)
freqs = FFTW.fftshift(freqs)

figure(figsize=(12,3))

subplot(1,2,1)
plot(freqs,abs.(audio_fft))
xlim([-2000,2000])
xlabel("Frequency in Hz")
ylabel("Magnitude")
title("FFT-Spectrum")

subplot(1,2,2)
plot_spectrogram(audio, Fs)
ylim([0,8000])
title("STFT-Spectrogram");
```

### House/3M Single Channel CI [fretz+1985]

![title](img/House_3M_Single_Channel_CI.png)


```julia
responsetype = Bandpass(340,2700; fs=Fs);
prototype = Butterworth(8);
cochlear_filter = digitalfilter(responsetype, prototype);

 = 0:0.01:pi 
H = freqz(cochlear_filter, )

figure(figsize=(12,3))

suptitle("Bandpass-Filter")
subplot(1,2,1)
plot(Fs/2*/pi, 20*log10.(abs.(H)))
xlabel("Frequency [Hz]")
ylabel("Gain [dB]")
xlim([0,last(Fs/2*/pi)])
grid()

subplot(1,2,2)
plot(Fs/2*/pi, 20*log10.(abs.(H)))
xlabel("Frequency [Hz]")
ylabel("Gain in dB")
xlim([0,3500])
ylim([-60,5])
grid()

# Filtering the input signal
audio_filt = filt(cochlear_filter, audio);
```

```julia
figure(figsize=(12,3))
subplot(1,2,1)
plot_spectrogram(audio, Fs)
ylim([0,5000])
title("Original Audio")

subplot(1,2,2)
plot_spectrogram(audio_filt, Fs)
ylim([0,5000])
title("Filtered Audio");
```

```julia
f_carrier(f,t) = sin.(2*pi*f*t)

f_carrier_16kHz = f_carrier(16e3,time) 
f_electrode = f_carrier_16kHz .* audio_filt;
```

```julia
subplot(3,1,1)
plot(time, f_carrier_16kHz)
xlim([first(time),last(time)])
xlim([0.1,0.105])
xlabel("Time in s") 
ylabel("Amplitude") 
title("Carrier signal")

subplot(3,1,2)
plot(time, audio_filt)
xlim([0.2,0.205])
ylim([-0.5,0.5])
xlabel("Time in s") 
ylabel("Amplitude") 
title("Filtered Audio Signal")

subplot(3,1,3)
plot(time, f_electrode)
xlim([0.2,0.205])
ylim([-0.5,0.5])
xlabel("Time in s") 
ylabel("Amplitude") 
tight_layout()
title("Electrode Signal");
```

```julia
figure(figsize=(12,3))
subplot(1,2,1)
plot_spectrogram(audio_filt, Fs)
ylim([0,5000])
title("Filtered Audio")

subplot(1,2,2)
plot_spectrogram(f_electrode, Fs)
#ylim([0,5000])
title("Electrode Signal");
```

### Multi-channel Analog CI

![title](img/Multi_channel_Analog_CI.png)


```julia
freq_passband = [[0.1,0.7]   [0.7,1.4]    [1.4,2.3]   [2.3, 5]]*1e3

filter_output = zeros(4,length(audio))
 = 0:0.01:pi 

legend_str = Array{Union{Nothing, String}}(nothing, 4)
for i in 1:4
    responsetype = Bandpass(freq_passband[1,i], freq_passband[2,i]; fs=Fs);
    prototype = Butterworth(8);
    cochlear_filter = digitalfilter(responsetype, prototype);
    
    H = freqz(cochlear_filter, )
    plot(Fs/2*/pi, 20*log10.(abs.(H)))
    xlim([0,10_000])
    ylim([-100,10])
    xlabel("Frequency in Hz")
    ylabel("Gain in dB")
    filter_output[i,:] = filt(cochlear_filter, audio);
    legend_str[i] = string("H",i,": ","( ",freq_passband[1,i]," - ",freq_passband[2,i]," ) Hz")
end
legend(legend_str)
```

### Feature Extraction Strategy F0-F2

![title](img/feature_f0_f2.png)

```julia
filter_output = zeros(2,length(audio))
 = 0:0.01:pi 

prototype = Butterworth(16);

responsetype = Lowpass(270; fs=Fs);
cochlear_filter = digitalfilter(responsetype, prototype);
H_low = freqz(cochlear_filter, )
filter_output[1,:] = filt(cochlear_filter, audio);

responsetype = Bandpass(1e3, 4e3; fs=Fs);
cochlear_filter = digitalfilter(responsetype, prototype);
H_bp = freqz(cochlear_filter, )
filter_output[2,:] = filt(cochlear_filter, audio);

plot(Fs/2*/pi, 20*log10.(abs.(H_low)))
plot(Fs/2*/pi, 20*log10.(abs.(H_bp)))
xlim([0,6_000])
ylim([-100,10])
xlabel("Frequency in Hz")
ylabel("Gain in dB")
legend(("H1: 270 Hz LP","H2: 1-4 kHz BP"));
```

```julia
figure(figsize=(12,5))
title_str = ["270 Hz LP", "1-4 kHz BP"]

for i in 1:2
    subplot(2,2,i)
    plot(time,filter_output[i,:])
    xlim([0,last(time)])
    xlabel("Time in s") 
    ylabel("Amplitude")
    title(title_str[i])
    
    subplot(2,2,i+2)
    plot(time,filter_output[i,:])
    xlim([0.2,0.25])
    xlabel("Time in s") 
    ylabel("Amplitude")
    grid()
end

tight_layout()

figure(figsize=(12,5))
subplot(2,1,1)
plot_spectrogram(filter_output[1,:], Fs)
ylim([0,1000])
title(title_str[1])
subplot(2,1,2)
plot_spectrogram(filter_output[2,:], Fs)
ylim([800,5000])
title(title_str[2])

tight_layout()
```

```julia
prototype = Butterworth(12);
responsetype = Lowpass(35; fs=Fs);
lp_filter = digitalfilter(responsetype, prototype);
rect_audio = abs.(filter_output[2,:])
rect_low = filt(lp_filter, rect_audio);

plot(time,rect_audio)
plot(time,rect_low,lw=2)
xlabel("Time in s") 
ylabel("Amplitude")
xlim([0, last(time)])
legend(["Rectified", "Low-pass filtered"]);
```

```julia
"""returns the number of times that the value changes its sign"""
function zero_crossings(array::AbstractVector{T}, size::Int = length(array), offset::Int = 0) where {T<:Real}
    result = 0
    previous = 0
    for i in offset .+ (1:size)
        number = array[i]
        sgn = number == 0 ? 0 : number > 0 ? 1 : -1
        if sgn != previous
            result += 1
        end
        previous = sgn
    end
    result
end;
```

```julia
function generate_frames(signal,Fs)
    
    frame_size = 25e-3
    frame_stride = 10e-3

    frame_length, frame_step = frame_size * Fs, frame_stride * Fs  # Convert from seconds to samples

    signal_length = length(signal)
    frame_length = convert(Int, round(frame_length))
    frame_step = convert(Int, round(frame_step))
    num_frames = convert(Int, ceil(convert(Float64, abs.(signal_length - frame_length)) / frame_step))  # Make sure that we have at least 1 frame

    pad_signal_length = num_frames * frame_step + frame_length
    z = zeros((pad_signal_length - signal_length))
    pad_signal = []
    append!(pad_signal, signal, z)

    indices1 = transpose( repeat(collect(0 : frame_length-1), outer = [1,num_frames]) )
    indices2 = repeat(collect(1 : frame_step : num_frames * frame_step), outer=[1,frame_length] )
    indices = indices1 + indices2
    frames = pad_signal[indices]
    return frames
    
end;
```

```julia
signal_frames_F0 = generate_frames(filter_output[1,:],Fs);
signal_frames_F2 = generate_frames(filter_output[2,:],Fs);

figure(figsize=(12,5))
subplot(2,2,1)
plot(signal_frames_F0[40,:])
legend(["Frame 40"])
xlabel("Samples")
ylabel("Amplitude")
title(title_str[1])
subplot(2,2,2)
plot(signal_frames_F2[40,:])
legend(["Frame 40"])
xlabel("Samples")
ylabel("Amplitude")
title(title_str[2])

F0_ZC = zeros((size(signal_frames_F0)[1],1))
F2_ZC = zeros((size(signal_frames_F2)[1],1))

for (idx,i) in enumerate(1:(size(signal_frames_F0)[1]))
    ZCR_F0 = zero_crossings(convert(Array{Float64,1}, signal_frames_F0[i,:]))
    F0_ZC[idx] = ZCR_F0/length(signal_frames_F0[i,:])*Fs/2
    ZCR_F2 = zero_crossings(convert(Array{Float64,1}, signal_frames_F2[i,:]))
    F2_ZC[idx] = ZCR_F2/length(signal_frames_F2[i,:])*Fs/2
end

subplot(2,2,3)
plot(time_frames,F0_ZC)
xlabel("Time in s")
ylabel("Frequency in Hz")
title("Zero-Crossing Frequency Estimation")
subplot(2,2,4)
plot(time_frames,F2_ZC)
xlabel("Time in s")
ylabel("Frequency in Hz")
title("Zero-Crossing Frequency Estimation")
tight_layout()
```
